```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
purchase_data
```

## Player Count

* Display the total number of players


```python
#drop duplicate names from ['SN'] 
total_players = purchase_data.drop_duplicates(subset=['SN'])
#total_players
#count number of values in 'SN' now that duplicates are gone
player_total = total_players['SN'].count()
player_total
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#number of unique items, average price of an item, total revenue, number of purchases
#create data frame showing all unique items
unique_items = purchase_data.drop_duplicates(subset=['Item Name'])
number_of_unique_items = unique_items['Purchase ID'].count()

#find average price from unique items
average_price = unique_items['Price'].mean()
average_price = round(average_price, 2)
#average_price

#find total revenue
total_revenue = purchase_data['Price'].sum()
#total_revenue

#find number of purchases
number_of_purchases = purchase_data['Purchase ID'].count()
#number_of_purchases

#create dictionary with these values
pymoli_data_dict = {'Number of Unique Items' : number_of_unique_items,
                               'Average Price' : average_price,
                               'Number of Purchases' : number_of_purchases,
                               'Total Revenue' : total_revenue}
#create summary data frame
pymoli_summary = pd.DataFrame(pymoli_data_dict, index=[0])
pymoli_summary

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins and labels
age_bins=[0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 150]
bin_labels=['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#age_demo['Total Count'] = pd.cut(total_players['Age'], bins=age_bins, labels=bin_labels)
age_count = pd.cut(total_players['Age'], bins=age_bins, labels=bin_labels).value_counts()

age_demo= pd.DataFrame()
age_demo['Total Count']= age_count
age_demo['Percentage of Players']= ((age_count/player_total)*100).map("{:,.2f}%".format)
age_demo = age_demo.sort_index()
age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#bin calls cut() --- 3 args, 1st is the df, 2nd is the range of values, 3rd is a list of the name sof the bins

#make age range column on original df
purchase_data['Age Ranges'] = pd.cut(purchase_data['Age'], age_bins, labels=bin_labels)
#purchase_data

#group by the new column: Age Ranges
group_data = purchase_data.groupby('Age Ranges')

#Collect values for Purchase Count relative to age groups(bins) using count()
count = group_data.count()['SN']
#print(count)

#collect values for average purchase price using .mean()
average_pprice = group_data.mean()['Price']
#print(average_pprice)

#collect values for total purchase value using .sum()
total_pvalue= group_data.sum()['Price']
#print(total_pvalue)

#calculate avg total purchase per person by dividing total purchase value by number of players in each demographic
avg_total_p = total_pvalue/age_count
#print(avg_total_p)

purchasing_analysis = pd.DataFrame()
purchasing_analysis['Purchase Count'] = count
purchasing_analysis['Average Purchase Price'] = average_pprice.map("${:,.2f}".format)
purchasing_analysis['Total Purchase Value'] = total_pvalue.map("${:,.2f}".format)
purchasing_analysis['Avg Total Purchase per Person'] = avg_total_p.map("${:,.2f}".format)
purchasing_analysis
```
